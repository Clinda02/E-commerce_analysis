# (2) 전처리

```python
# Order ID의 결측치 확인
Olist[Olist['Order ID'].isna()]
```

```python
# 각 컬럼의 결측치 확인
Olist[Olist['Order ID'].isna()].isnull().sum()
```

```python
# 결측치 제거
Olist = Olist[Olist['Order ID'].isna()==False]
Olist.shape
```

```python
# Olist, Detail 데이터 병합
data = Olist.merge(Detail, on = 'Order ID')
```

```python
data.shape
```

```python
# merge 후 데이터 누락 및 중복 확인
print(f"Olist의 'Order ID' 개수 : {len(Olist['Order ID'].unique())}")
print(f"data의 'Order ID' 개수 : {len(data['Order ID'].unique())}")
print(f"Detail의 'Order ID' 개수 : {len(Detail['Order ID'].unique())}")
```

```python
# 결측치 확인
data.isnull().values.any()
```

```python
data.head()
```

```python
# 날짜 변환
data['Order Date'] = pd.to_datetime(data['Order Date'], format = '%d-%m-%Y')
```

```python
# 주문 시작일, 최근 주문일 조회
data['Order Date'].min(), data['Order Date'].max()
```

```python
# 연, 월, 연월 칼럼 추가
data['year'] = data['Order Date'].dt.year
data['month'] = data['Order Date'].dt.month
data['yearmonth'] = data['Order Date'].astype('str').str.slice(0,7)
```

```python
data.head(1)
```

# (3) plotly를 이용한 EDA 시각화

```python
# 라이브러리 임포트
import plotly.express as px
```

```python
# 연월별 전체 주문수량 집계
df = data.groupby('yearmonth').agg({'Quantity' : 'sum'}).reset_index()
df
```

### Line 그래프 그리기

```python
# 연월별 전체 주문수량 라인그래프 생성
fig1 = px.line(df, x='yearmonth', y="Quantity", title='Sales Quantity by month')
fig1.show()
```

```python
# 연월별 전체 판매금액(매출) 집계 및 시각화
df2 = data.groupby('yearmonth').agg({'Amount' : 'sum'}).reset_index()
fig2 = px.line(df2, x='yearmonth', y="Amount", title='Sales Amount by month')
fig2.show()
```

### Bar 그래프 그리기

```python
# 카테고리별 판매수량 집계
df3 = data.pivot_table(index = 'Category', values=['Quantity'], aggfunc = 'sum').reset_index()
df3
```

```python
# 카테고리별 판매수량 바그래프 생성
fig3 = px.bar(df3, x = 'Category', y = 'Quantity')
fig3.show()
```

```python
# 월별로 카테고리 주문수량 집계
df4 = data.pivot_table(index = ['yearmonth','Category'], values='Quantity', aggfunc = 'sum').reset_index()
df4.head()
```

* 누적 Bar 그래프

```python
# 월별로 카테고리 주문수량 누적그래프 생성
fig4 = px.bar(df4, x = 'yearmonth', y = 'Quantity', color = 'Category')
fig4.show()
```

* 전체 주문수량이 증가하게 된 주요 요인은 **Clothing** 카테고리다.

### heat map을 이용한 지역별 주력 판매상품 분석

```python
# 지역별 세부 카테고리의 주문수량/주문금액/매출량 집계
df5 = data.pivot_table(index = ['State','Sub-Category'], values = ['Quantity', 'Amount', 'Profit'], aggfunc = 'sum').reset_index()
df5.head()
```

```python
# 밀도에 따른 히트맵 생성
fig5 = px.density_heatmap(df5, x = 'State', y = 'Sub-Category', z = 'Quantity', title='Quantity heat map') # z : 밀도
fig5.show()

fig6 = px.density_heatmap(df5, x = 'State', y = 'Sub-Category', z = 'Amount', title='Amount heat map')
fig6.show()
```
